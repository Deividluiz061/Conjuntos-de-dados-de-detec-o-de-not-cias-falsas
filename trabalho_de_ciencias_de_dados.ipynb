```python
# prompt: importe o pandas matplotib

# Importando a biblioteca pandas para trabalhar com DataFrames (tabelas de dados)
import pandas as pd
# Importando a biblioteca matplotlib.pyplot para criar gráficos
import matplotlib.pyplot as plt
```

```python
# prompt: trabalha co arquivo que esta nesse caminho /fake_news_dataset.csv

# Carregando o conjunto de dados de um arquivo CSV para um DataFrame do pandas
# O arquivo CSV está localizado no caminho '/fake_news_dataset.csv'
df = pd.read_csv('/fake_news_dataset.csv')

# Exibindo as primeiras 5 linhas do DataFrame para ter uma ideia dos dados
print("As primeiras 5 linhas do conjunto de dados:")
display(df.head())
```

```python
# 1. Quais são as colunas presentes no conjunto de dados e quais são os tipos de dados de cada coluna?
print("Informações sobre as colunas e tipos de dados:")
# O método .info() fornece um resumo do DataFrame, incluindo nomes das colunas,
# contagem de valores não nulos e tipos de dados.
display(df.info())
```

```python
# 2. Quantas linhas e colunas o conjunto de dados possui?
print("\nFormato do conjunto de dados (linhas, colunas):")
# O atributo .shape retorna uma tupla com o número de linhas e colunas do DataFrame.
display(df.shape)
```

```python
# 3. Existem valores ausentes em alguma coluna? Se sim, em quais e quantos?
print("\nValores ausentes por coluna:")
# .isnull() cria um DataFrame booleano (True onde o valor é nulo)
# .sum() soma os valores True por coluna, contando os valores ausentes.
display(df.isnull().sum())
```

```python
# 4. Quais são os valores únicos na coluna 'label' e qual a distribuição de cada valor?
print("\nDistribuição dos valores na coluna 'label':")
# .value_counts() conta a frequência de cada valor único em uma coluna.
display(df['label'].value_counts())
```

```python
# 5. Quais são as fontes ('source') mais frequentes no conjunto de dados?
print("\nFontes mais frequentes:")
# .value_counts() conta a frequência de cada fonte.
# .head() mostra apenas as 5 primeiras (mais frequentes).
display(df['source'].value_counts().head())
```

```python
# 1. Gráfico de barras da distribuição de 'label'
# Criando uma figura para o gráfico com um tamanho específico
plt.figure(figsize=(6, 4))
# Contando os valores na coluna 'label' e plotando um gráfico de barras
df['label'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
# Definindo o título do gráfico
plt.title('Distribuição de Notícias Falsas vs. Verdadeiras')
# Definindo o rótulo do eixo X
plt.xlabel('Rótulo')
# Definindo o rótulo do eixo Y
plt.ylabel('Contagem')
# Removendo a rotação dos rótulos do eixo X para melhor legibilidade
plt.xticks(rotation=0)
# Mostrando o gráfico
plt.show()

print("Este gráfico de barras mostra a contagem de notícias rotuladas como 'fake' (falsas) e 'real' (verdadeiras). É útil para ver se o conjunto de dados está equilibrado em termos de notícias falsas e verdadeiras.")
```

```python
# 2. Gráfico de barras das fontes mais frequentes
# Criando uma figura para o gráfico
plt.figure(figsize=(10, 6))
# Contando as 10 fontes mais frequentes e plotando um gráfico de barras
df['source'].value_counts().head(10).plot(kind='bar', color='lightgreen')
# Definindo título e rótulos dos eixos
plt.title('Top 10 Fontes de Notícias Mais Frequentes')
plt.xlabel('Fonte')
plt.ylabel('Contagem')
# Rotacionando os rótulos do eixo X para evitar sobreposição
plt.xticks(rotation=45, ha='right')
# Ajustando o layout para evitar cortes nos rótulos
plt.tight_layout()
# Mostrando o gráfico
plt.show()

print("Este gráfico de barras exibe as 10 fontes de notícias que aparecem com mais frequência no conjunto de dados. Ajuda a identificar quais fontes são mais representadas.")
```

```python
# 3. Gráfico de barras da distribuição de 'category'
# Criando uma figura para o gráfico
plt.figure(figsize=(10, 6))
# Contando os valores na coluna 'category' e plotando um gráfico de barras
df['category'].value_counts().plot(kind='bar', color='gold')
# Definindo título e rótulos dos eixos
plt.title('Distribuição de Notícias por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Contagem')
# Rotacionando os rótulos do eixo X
plt.xticks(rotation=45, ha='right')
# Ajustando o layout
plt.tight_layout()
# Mostrando o gráfico
plt.show()

print("Este gráfico de barras mostra a distribuição das notícias por diferentes categorias (como Política, Esporte, etc.). Permite ver quais categorias são mais comuns no conjunto de dados.")
```

```python
# 4. Gráfico de barras de valores ausentes por coluna
# Contando os valores ausentes em cada coluna
missing_values = df.isnull().sum()
# Filtrando para mostrar apenas as colunas que têm valores ausentes (contagem > 0)
missing_values = missing_values[missing_values > 0]

# Verificando se há colunas com valores ausentes antes de tentar plotar
if not missing_values.empty:
    # Criando uma figura para o gráfico
    plt.figure(figsize=(8, 5))
    # Plotando um gráfico de barras com a contagem de valores ausentes por coluna
    missing_values.plot(kind='bar', color='coral')
    # Definindo título e rótulos dos eixos
    plt.title('Número de Valores Ausentes por Coluna')
    plt.xlabel('Coluna')
    plt.ylabel('Contagem de Valores Ausentes')
    # Rotacionando os rótulos do eixo X
    plt.xticks(rotation=45, ha='right')
    # Ajustando o layout
    plt.tight_layout()
    # Mostrando o gráfico
    plt.show()
    print("Este gráfico de barras visualiza o número de valores ausentes em cada coluna que possui dados faltando. É útil para identificar rapidamente quais colunas precisam de tratamento para dados ausentes.")
else:
    print("Não há valores ausentes no conjunto de dados para gerar este gráfico.")
```

```python
# 5. Gráfico de barras da distribuição de 'label' por 'source' (para as 5 fontes mais frequentes)
# Primeiro, vamos obter as 5 fontes mais frequentes
top_sources = df['source'].value_counts().head(5).index.tolist()

# Filtrar o dataframe para incluir apenas essas fontes para esta análise
df_top_sources = df[df['source'].isin(top_sources)]

# Criar uma tabela de contingência (crosstab) que mostra a contagem de 'fake' e 'real'
# para cada uma das 5 fontes mais frequentes.
contingency_table = pd.crosstab(df_top_sources['source'], df_top_sources['label'])

# Plotar o gráfico de barras empilhadas a partir da tabela de contingência
contingency_table.plot(kind='bar', stacked=True, figsize=(10, 6), color=['salmon', 'skyblue'])
# Definindo título e rótulos dos eixos
plt.title('Distribuição de Rótulos (Fake/Real) nas 5 Fontes Mais Frequentes')
plt.xlabel('Fonte')
plt.ylabel('Contagem')
# Rotacionando os rótulos do eixo X
plt.xticks(rotation=45, ha='right')
# Adicionando uma legenda para identificar as barras 'fake' e 'real'
plt.legend(title='Rótulo')
# Ajustando o layout
plt.tight_layout()
# Mostrando o gráfico
plt.show()

print("Este gráfico de barras empilhadas compara a distribuição de notícias falsas e verdadeiras dentro das 5 fontes de notícias mais frequentes. Ele nos ajuda a ver se certas fontes tendem a ter mais notícias falsas ou verdadeiras do que outras.")
```
